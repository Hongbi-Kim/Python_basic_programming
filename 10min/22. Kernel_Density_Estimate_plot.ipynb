## 정규분포
- 확률론과 통계학에서, 정규 분포(normal distribution) 또는 가우시안 분포(Gaussian distribution)는 연속 확률 분포의 하나이다.
- 정규분포를 근사하는데에 자주 사용되며, 이것은 중심극한정리에 의하여 독립적인 확률변수들의 평균은 정규분포에 가까워지는 성질

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline

# 한글폰트 사용 시 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
mpl.rcParams['axes.unicode_minus'] = False
```

```python
ser = pd.Series(np.random.randn(1000)) # 1000개의 데이터 생성(무작위)
ser.plot.hist()
```

```python
ser.plot.density() # 정규분포 형태의 확률밀도함수
```

```python
ser.plot.kde()
```
