## Pie Plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline

# 한글폰트 사용 시 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
mpl.rcParams['axes.unicode_minus'] = False
```

```python
series = pd.Series(np.random.rand(4),
                  index = ['a','b','c','d'], name = 'series')
series
```

```python
series = pd.Series(3*np.random.rand(4),
                  index = ['a','b','c','d'], name = 'series')
```

```python
series
```

```python
series.plot.pie()
```

```python
series.plot.pie(figsize = (6,6))
```

```python
series.plot.pie(labels = ['AA','BB','CC','DD'],
               colors = ['r','g','b','c'],
               autopct = '%.2f',
               fontsize = 20,
               figsize = (6,6))
```

```python
df = pd.DataFrame({'mass':[0.330, 4.87, 5.97],
                  'radius' : [2439.7, 6051.8, 6378.1]},
                 index = ['Mercury', 'Venus','Earth'])
```

```python
df
```

```python
plot = df.plot.pie(figsize = (10,5))
```

```python
df_m = df.mass # df['mass']
plot = df_m.plot.pie(figsize = (10,5))
```

```python
plot = df.plot.pie(subplots = True, figsize = (10,5))
```
